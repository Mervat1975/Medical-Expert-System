```python
def perform_diagnosis(personal_info, symptom_list):
	header()    
	preprocess()
	#engine = MedicalChatbot(personal_info, symptom_list)
	engine = MedicalChatbot()
	engine.set_params(personal_info, symptom_list)
	continue_flag=1
	engine.reset()  # Prepare the engine for the execution.
	engine.run()  # Run it!
	#print(engine.facts)
```

```python
# generate layout

def generate_layout():
    # left column 
    name = widgets.Text(
        placeholder='Enter a name',
        description='Name:',
    )
    age = widgets.Text(
        placeholder='Enter a number',
        description='Age:',
    )
    gender = widgets.RadioButtons(
        options=['Male', 'Female'],
        description='Gender',
        disabled=False
    )

    left_box = widgets.VBox(children=[name, age, gender])

    # right column
    data = ["chest pain", "Nausea", "low body temperature", "sunken eyes", "sore throat",
           "back pain", "fever", "cough", "fatigue", "headache", "restlessness", "blurred vision", "fainting"]
    checkboxes = [widgets.Checkbox(value=False, description=label) for label in data]
    right_box = widgets.VBox(children=checkboxes)
    
    button = widgets.Button(description="Click to Diagnose")
    output = widgets.Output(layout={'border': '1px solid black'})

    
    third_column = widgets.VBox(children=[button])

    def on_button_clicked(b):
        output.clear_output()
        with output:
            symptoms_list = ['no' if check.value is False else 'yes' for check in checkboxes]
            personal_info = {'name': name.value, 'age': age.value, 'gender': gender.value}
            
            return perform_diagnosis(symptoms_list,personal_info)

    button.on_click(on_button_clicked)
    third_column = widgets.VBox(children=[button, output])
    
    Hbox = widgets.HBox([left_box, right_box,third_column], layout={'border': '2px solid black', 'background-color':'blue'})

    # Click button
    display(Hbox)
    
generate_layout()
```
